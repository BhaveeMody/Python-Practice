## factorial

```python
a = int(input("enter any number."))
f=1
i=a
while a>=1:
  f=f*a
  a=a-1
print("factorial of",i,"=",f)
```

## Prime

```python
a = int(input("enter any number"))
p=1
for i in range(2,a-1):
  if a%i==0:
    p=0
if p==0:
  print("This number is not a prime number.")
else:
  print("This number is a prime number.")

```

## Fibonacci Series First 10


```python
num1=0
num2=1
print(num1)
print(num2)
i=1
while i<9:
  num3=num1+num2
  print(num3)
  num1=num2
  num2=num3
  i=i+1
```

```python
n = int(input("Enter the number of terms: "))
a, b = 0, 1
count = 0
while count < n:
    print(a, end=" ")
    a, b = b, a + b
    count += 1
```

```python
print("Pattern: P")
for row in range(7):
    for col in range(5):
        if col == 0 or ((row == 0 or row == 3) and col < 4) or (col == 4 and (row != 0 and row != 3 and row<3)):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
print("Pattern: C")
for row in range(7):
    for col in range(5):
        if (col == 0) or(row == 0) or (row == 6):
            print("*", end=" ")
        elif (row == 0 and col == 0) or (row == 6 and col == 0):
            print(" ",end=" ")
        else:
            print(" ", end=" ")
    print()
```
